```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/data.csv')
```

```python
data.head()
```

```python
current_day_of_progress = data["current day"]
week_day = data["week day"]
month = data["month"]
year = data["year"]
day = data["day"]
protein = data["protein"]
carbs = data["carbs"]
fat = data["fat"]
calories_intake = data["calories intake"]
dwater = data["water"]
move_cal = data["move cal"]
type_of_training = data["training day"]
```

```python
# data.sort_values(by=['move cal'])
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.bar(current_day_of_progress, move_cal, width=1, edgecolor="black", color="grey", linewidth=0.7)

font = {
    'size': 18,
    'fontweight': 'bold'
}

title = {
    'size': 23,
    'fontweight': 'bold',
}

# set daily goal
plt.axhline(y = 500, color = 'r', linestyle = '--')

plt.title('Moving calories', fontdict = title)
plt.ylabel('move cal', fontdict = font) 
plt.xlabel('current days', fontdict = font)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,8))
x = np.arange(len(current_day_of_progress))


width = 0.2
plt.bar(x-0.2, protein,
        width, color='red', label='protein')
plt.bar(x, carbs,
        width, color='green', label='carbs')
plt.bar(x+0.2, fat,
        width, color='gold', label='fat')

plt.axhline(y = 73, color='gold', linestyle = '--', label='73')
plt.axhline(y = 142, color='green', linestyle = '--', label='142')
plt.axhline(y = 165, color='red', linestyle = '--', label='165')

plt.title('Macros', fontdict = title)
plt.legend(fontsize=15)
plt.ylabel('gramms per day', fontdict = font)
plt.show()
```
